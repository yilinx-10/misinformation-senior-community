```python
import random
import seaborn as sns
import matplotlib.pyplot as plt

# Parameters for beta distribution
alpha_a, beta_a = 3, 5  # Skewed to the left (more low values)
alpha_b, beta_b = 5, 3  # Skewed to the right (more high values)

# Generate samples
samples_a = [random.betavariate(alpha_a, beta_a) for _ in range(1000)]
samples_b = [random.betavariate(alpha_b, beta_b) for _ in range(1000)]

# Plot KDE
plt.figure(figsize=(10, 6))
sns.kdeplot(samples_a, label=f"Entity A (={alpha_a}, ={beta_a})", fill=True, color="blue", alpha=0.4)
sns.kdeplot(samples_b, label=f"Entity B (={alpha_b}, ={beta_b})", fill=True, color="orange", alpha=0.4)

plt.title("KDE Plot of Beta-Distributed Samples")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta

# Define the x-axis range
x = np.linspace(0, 1, 500)

# List of a values to iterate through
a_values = [1, 2, 3, 4, 5]

# Set up a color map
colors = plt.cm.viridis(np.linspace(0, 1, len(a_values)))  # 5 distinct colors

# Create the plot
plt.figure(figsize=(12, 6))

# Loop through a_values and assign matching colors
for i, a in enumerate(a_values):
    color = colors[i]
    
    # Plot Beta(a, 5) with solid line
    plt.plot(x, beta.pdf(x, a, 5), linestyle='-', color=color, label=f'Beta({a}, 5)')
    
    # Plot Beta(5, a) with dashed line
    plt.plot(x, beta.pdf(x, 5, a), linestyle='--', color=color, label=f'Beta(5, {a})')

# Add title and labels with alpha in LaTeX
plt.title('Beta Distributions: $\\mathrm{Beta}(\\alpha, 5)$ (solid) and $\\mathrm{Beta}(5, \\alpha)$ (dashed)')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
